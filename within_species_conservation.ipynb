```python
import numpy as np
from scipy.spatial.distance import euclidean, hamming
import pandas as pd
from Bio import SeqIO
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shapes_path = "/Users/tennisnyjmac/Desktop/Promoters/dna_shape/athaliana_Opening_200.txt"

with open(shapes_path, 'r') as file:
    raw_lines = file.readlines()
                
    shapes = []
    for line in raw_lines:
        values = line.strip().split()
        if len(values) > 401:
            values = values[:401] 
        elif len(values) < 401:
            values += [np.nan] * (401 - len(values)) 
        shapes.append(values)


fasta_sequences = list(SeqIO.parse("/Users/tennisnyjmac/Desktop/Promoters/athaliana_200.fa", "fasta"))
```

```python
sample_indices = random.sample(range(len(fasta_sequences)), 100)
sampled_fasta_sequences = [fasta_sequences[i] for i in sample_indices]
sampled_shapes = [shapes[i] for i in sample_indices]

sequence_distances = []
shape_distances = []

for i in range(len(sampled_fasta_sequences)):
    for j in range(i + 1, len(sampled_fasta_sequences)):
        seq1 = sampled_fasta_sequences[i].seq
        seq2 = sampled_fasta_sequences[j].seq
        seq_dist = hamming(list(seq1), list(seq2)) * len(seq1)
        sequence_distances.append(seq_dist)

        shape1 = np.array(sampled_shapes[i], dtype=float)
        shape2 = np.array(sampled_shapes[j], dtype=float)

        if not np.any(np.isnan(shape1)) and not np.any(np.isnan(shape2)):
            shape_dist = euclidean(shape1, shape2)
            shape_distances.append(shape_dist)
```

```python
sequence_distances = np.array(sequence_distances)
shape_distances = np.array(shape_distances)

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(sequence_distances, kde=True, color='blue')
plt.title('Distribution of Sequence Distances')
plt.xlabel('Hamming Distance')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(shape_distances, kde=True, color='green')
plt.title('Distribution of Shape Distances')
plt.xlabel('Euclidean Distance')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
normalized_sequence_distances = (sequence_distances - np.mean(sequence_distances)) / np.std(sequence_distances)
normalized_shape_distances = (shape_distances - np.mean(shape_distances)) / np.std(shape_distances)

# Plot the normalized distributions
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(normalized_sequence_distances, kde=True, color='blue')
plt.title('Normalized Distribution of Sequence Distances')
plt.xlabel('Normalized Hamming Distance')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(normalized_shape_distances, kde=True, color='green')
plt.title('Normalized Distribution of Shape Distances')
plt.xlabel('Normalized Euclidean Distance')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
